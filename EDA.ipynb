```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("viridis")



# 1. Distribution plots for numerical features
def plot_numerical_distributions(df):
    numerical_features = ['RegLen', 'AgeofDomain', 'WebsiteTraffic', 'PageRank', 'GoogleIndex', 'LinksPointingToPage']

    # Filter only numerical features that exist in the dataframe
    numerical_features = [feat for feat in numerical_features if feat in df.columns]

    if len(numerical_features) == 0:
        return None

    fig, axes = plt.subplots(len(numerical_features), 1, figsize=(12, 4*len(numerical_features)))

    if len(numerical_features) == 1:
        axes = [axes]

    for i, feature in enumerate(numerical_features):
        sns.histplot(df[feature], kde=True, ax=axes[i])
        axes[i].set_title(f'Distribution of {feature}')
        axes[i].set_xlabel(feature)
        axes[i].set_ylabel('Frequency')

    plt.tight_layout()
    return fig

# 2. Box plots for comparing features between classes
def plot_boxplots(df):
    numerical_features = ['RegLen', 'AgeofDomain', 'WebsiteTraffic', 'PageRank', 'GoogleIndex', 'LinksPointingToPage']


    numerical_features = [feat for feat in numerical_features if feat in df.columns]

    if len(numerical_features) == 0 or 'class' not in df.columns:
        return None

    fig, axes = plt.subplots(len(numerical_features), 1, figsize=(12, 4*len(numerical_features)))

    if len(numerical_features) == 1:
        axes = [axes]

    for i, feature in enumerate(numerical_features):
        sns.boxplot(x='class', y=feature, data=df, ax=axes[i])
        axes[i].set_title(f'Box Plot of {feature} by Class')
        axes[i].set_xlabel('Class (0: Legitimate, 1: Phishing)')
        axes[i].set_ylabel(feature)

    plt.tight_layout()
    return fig

# 3. Correlation heatmap
def plot_correlation_heatmap(df):
    numerical_df = df.select_dtypes(include=[np.number])
    corr_matrix = numerical_df.corr()

    # Create heatmap
    plt.figure(figsize=(12, 10))
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
                linewidths=0.5, cbar_kws={'shrink': .8})
    plt.title('Correlation Heatmap of Numerical Features')
    plt.tight_layout()
    return plt.gcf()

# 4. Scatter plot with domain age vs PageRank
def plot_domain_age_vs_pagerank(df):
    if 'AgeofDomain' not in df.columns or 'PageRank' not in df.columns or 'class' not in df.columns:
        return None

    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='AgeofDomain', y='PageRank', hue='class', data=df, palette='viridis', alpha=0.7)
    plt.title('Domain Age vs PageRank by Class')
    plt.xlabel('Age of Domain')
    plt.ylabel('PageRank')
    plt.legend(title='Class', labels=['Legitimate', 'Phishing'])
    return plt.gcf()

# 5. Parallel coordinates plot
def plot_parallel_coordinates(df):

    features = ['RegLen', 'AgeofDomain', 'WebsiteTraffic', 'PageRank', 'GoogleIndex', 'LinksPointingToPage']


    features = [feat for feat in features if feat in df.columns]

    if len(features) == 0 or 'class' not in df.columns:
        return None


    if len(df) > 1000:
        df_sample = df.sample(1000, random_state=42)
    else:
        df_sample = df

    # Create the plot
    fig = px.parallel_coordinates(
        df_sample,
        color='class',
        dimensions=features,
        color_continuous_scale=px.colors.diverging.Tealrose,
        title='Parallel Coordinates Plot of Features by Class'
    )
    return fig

# 6. Violin plots for comparing distributions
def plot_violin_plots(df):
    features = ['WebsiteTraffic', 'GoogleIndex']


    features = [feat for feat in features if feat in df.columns]

    if len(features) == 0 or 'class' not in df.columns:
        return None

    fig, axes = plt.subplots(1, len(features), figsize=(12, 6))

    if len(features) == 1:
        axes = [axes]

    for i, feature in enumerate(features):
        sns.violinplot(x='class', y=feature, data=df, ax=axes[i], inner='quartile')
        axes[i].set_title(f'Violin Plot of {feature} by Class')
        axes[i].set_xlabel('Class (0: Legitimate, 1: Phishing)')
        axes[i].set_ylabel(feature)

    plt.tight_layout()
    return fig


def plot_radar_chart(df):
    binary_features = ['HTTPS', 'NonStdPort', 'AbnormalURL', 'UsingPopupWindow',
                       'IframeRedirection', 'DisableRightClick', 'UsingIPLongURL']

    binary_features = [feat for feat in binary_features if feat in df.columns]

    if len(binary_features) == 0 or 'class' not in df.columns:
        return None


    class_0_means = df[df['class'] == 0][binary_features].mean().values
    class_1_means = df[df['class'] == 1][binary_features].mean().values

    # Create radar chart
    fig = go.Figure()

    fig.add_trace(go.Scatterpolar(
        r=class_0_means,
        theta=binary_features,
        fill='toself',
        name='Legitimate'
    ))

    fig.add_trace(go.Scatterpolar(
        r=class_1_means,
        theta=binary_features,
        fill='toself',
        name='Phishing'
    ))

    fig.update_layout(
        polar=dict(
            radialaxis=dict(
                visible=True,
                range=[0, 1]
            )
        ),
        title='Radar Chart of Binary Features by Class',
        showlegend=True
    )

    return fig


def plot_pair_plots(df):
    selected_features = ['RegLen', 'AgeofDomain', 'PageRank', 'GoogleIndex']


    selected_features = [feat for feat in selected_features if feat in df.columns]

    if len(selected_features) < 2 or 'class' not in df.columns:
        return None


    if len(df) > 1000:
        df_sample = df.sample(1000, random_state=42)
    else:
        df_sample = df


    pair_plot = sns.pairplot(df_sample, vars=selected_features, hue='class', diag_kind='kde',
                             plot_kws={'alpha': 0.6}, height=2.5)
    pair_plot.fig.suptitle('Pair Plot of Selected Features', y=1.02)
    return pair_plot.fig

def plot_dimension_reduction(df):

    numerical_df = df.select_dtypes(include=[np.number])

    if 'class' not in df.columns or len(numerical_df.columns) < 3:
        return None


    if 'class' in numerical_df.columns:
        features = numerical_df.drop('class', axis=1)
    else:
        features = numerical_df

    # Create a sample if the dataset is too large
    if len(df) > 5000:
        sample_idx = np.random.choice(len(df), 5000, replace=False)
        features_sample = features.iloc[sample_idx]
        classes_sample = df['class'].iloc[sample_idx]
    else:
        features_sample = features
        classes_sample = df['class']


    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features_sample)

    # Apply PCA
    pca = PCA(n_components=2)
    features_pca = pca.fit_transform(features_scaled)

    # Create PCA plot
    fig, axes = plt.subplots(1, 2, figsize=(18, 7))

    scatter_pca = axes[0].scatter(features_pca[:, 0], features_pca[:, 1], c=classes_sample,
                                  cmap='viridis', alpha=0.7, edgecolors='w', linewidth=0.5)
    axes[0].set_title('PCA Projection')
    axes[0].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)')
    axes[0].set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)')
    legend = axes[0].legend(*scatter_pca.legend_elements(), title="Classes")
    axes[0].add_artist(legend)


    tsne = TSNE(n_components=2, random_state=42)
    features_tsne = tsne.fit_transform(features_scaled)

    # Create t-SNE plot
    scatter_tsne = axes[1].scatter(features_tsne[:, 0], features_tsne[:, 1], c=classes_sample,
                                   cmap='viridis', alpha=0.7, edgecolors='w', linewidth=0.5)
    axes[1].set_title('t-SNE Projection')
    axes[1].set_xlabel('t-SNE Component 1')
    axes[1].set_ylabel('t-SNE Component 2')
    legend = axes[1].legend(*scatter_tsne.legend_elements(), title="Classes")
    axes[1].add_artist(legend)

    plt.tight_layout()
    return fig

# 10. Density plots
def plot_density_plots(df):
    if 'LinksPointingToPage' not in df.columns or 'class' not in df.columns:
        return None

    plt.figure(figsize=(12, 6))

    # Plot density for each class
    for class_value in df['class'].unique():
        sns.kdeplot(df[df['class'] == class_value]['LinksPointingToPage'],
                    label=f'Class {class_value}', fill=True, alpha=0.3)

    plt.title('Density Plot of Links Pointing to Page by Class')
    plt.xlabel('Links Pointing to Page')
    plt.ylabel('Density')
    plt.legend(title='Class', labels=['Legitimate', 'Phishing'])

    return plt.gcf()

# 11. Stacked bar charts
def plot_stacked_bars(df):
    binary_features = ['UsingPopupWindow', 'IframeRedirection', 'DisableRightClick', 'StatusBarCust']


    binary_features = [feat for feat in binary_features if feat in df.columns]

    if len(binary_features) == 0 or 'class' not in df.columns:
        return None

    # Create subplots
    fig, axes = plt.subplots(len(binary_features), 1, figsize=(10, 4*len(binary_features)))

    if len(binary_features) == 1:
        axes = [axes]

    for i, feature in enumerate(binary_features):

        props = df.groupby('class')[feature].value_counts(normalize=True).unstack().fillna(0)


        props.plot(kind='bar', stacked=True, ax=axes[i], color=['#3498db', '#e74c3c'])
        axes[i].set_title(f'Proportion of {feature} Values by Class')
        axes[i].set_xlabel('Class (0: Legitimate, 1: Phishing)')
        axes[i].set_ylabel('Proportion')
        axes[i].legend(title=feature, labels=['0 (No)', '1 (Yes)'])

    plt.tight_layout()
    return fig

# 12. Histogram overlays
def plot_histogram_overlays(df):
    numerical_features = ['RegLen', 'AgeofDomain', 'WebsiteTraffic', 'PageRank']


    numerical_features = [feat for feat in numerical_features if feat in df.columns]

    if len(numerical_features) == 0 or 'class' not in df.columns:
        return None

    fig, axes = plt.subplots(len(numerical_features), 1, figsize=(12, 4*len(numerical_features)))

    if len(numerical_features) == 1:
        axes = [axes]

    for i, feature in enumerate(numerical_features):
        for class_value in sorted(df['class'].unique()):
            sns.histplot(df[df['class'] == class_value][feature],
                         alpha=0.5,
                         label=f'Class {class_value}',
                         ax=axes[i],
                         kde=True)

        axes[i].set_title(f'Histogram of {feature} by Class')
        axes[i].set_xlabel(feature)
        axes[i].set_ylabel('Frequency')
        axes[i].legend(title='Class', labels=['Legitimate', 'Phishing'])

    plt.tight_layout()
    return fig


```

```python


# 1. Distribution plots for numerical features
fig1 = plot_numerical_distributions(df)
if fig1:
    plt.figure(fig1.number)
    plt.savefig('numerical_distributions.png', bbox_inches='tight')
    plt.show()

# 2. Box plots for comparing features between classes
fig2 = plot_boxplots(df)
if fig2:
    plt.figure(fig2.number)
    plt.savefig('boxplots.png', bbox_inches='tight')
    plt.show()

# 3. Correlation heatmap
fig3 = plot_correlation_heatmap(df)
if fig3:
    plt.figure(fig3.number)
    plt.savefig('correlation_heatmap.png', bbox_inches='tight')
    plt.show()

# 4. Scatter plot with domain age vs PageRank
fig4 = plot_domain_age_vs_pagerank(df)
if fig4:
    plt.figure(fig4.number)
    plt.savefig('domain_age_vs_pagerank.png', bbox_inches='tight')
    plt.show()

# 5. Parallel coordinates plot (Plotly)
fig5 = plot_parallel_coordinates(df)
if fig5:
    fig5.write_html('parallel_coordinates.html')

    # fig5.show()

# 6. Violin plots for comparing distributions
fig6 = plot_violin_plots(df)
if fig6:
    plt.figure(fig6.number)
    plt.savefig('violin_plots.png', bbox_inches='tight')
    plt.show()

# 7. Radar chart comparing binary features (Plotly)
fig7 = plot_radar_chart(df)
if fig7:
    fig7.write_html('radar_chart.html')


# 8. Pair plots for selected features
fig8 = plot_pair_plots(df)
if fig8:
    plt.figure(fig8.number)
    plt.savefig('pair_plots.png', bbox_inches='tight')
    plt.show()

# 9. TSNE or PCA visualization
fig9 = plot_dimension_reduction(df)
if fig9:
    plt.figure(fig9.number)
    plt.savefig('dimension_reduction.png', bbox_inches='tight')
    plt.show()

# 10. Density plots
fig10 = plot_density_plots(df)
if fig10:
    plt.figure(fig10.number)
    plt.savefig('density_plots.png', bbox_inches='tight')
    plt.show()

# 11. Stacked bar charts
fig11 = plot_stacked_bars(df)
if fig11:
    plt.figure(fig11.number)
    plt.savefig('stacked_bars.png', bbox_inches='tight')
    plt.show()

# 12. Histogram overlays
fig12 = plot_histogram_overlays(df)
if fig12:
    plt.figure(fig12.number)
    plt.savefig('histogram_overlays.png', bbox_inches='tight')
    plt.show()
```
